```python
plt.figure(figsize=(18, 18))
numeric_cols = olive_oil_df.select_dtypes(include=np.number).columns

for i, col in enumerate(numeric_cols):
    plt.subplot(5, 3, i+1) # each row three figure
    x = olive_oil_df[col] #x-axis
    y = olive_oil_df['Irrigation_L_per_tree'] #y-axis
    plt.plot(x, y, 'o')
    # Create regression line
    plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1)) (np.unique(x)),
    color='red')
    plt.xlabel(col) # x-label
    plt.ylabel('Irrigation_L_per_tree') # y-label
```

```python
plt.figure(figsize=(18, 18))
numeric_cols = olive_oil_df.select_dtypes(include=np.number).columns

for i, col in enumerate(numeric_cols):
    plt.subplot(5, 3, i+1) # each row three figure
    x = olive_oil_df[col] #x-axis
    y = olive_oil_df['Olives_Produced_kg_per_hectare'] #y-axis
    plt.plot(x, y, 'o')
    # Create regression line
    plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1)) (np.unique(x)),
    color='red')
    plt.xlabel(col) # x-label
    plt.ylabel('Olives_Produced_kg_per_hectare') # y-label
```

```python
# Create a new column for quantile bins based on 'Olives_Produced_kg_per_hectare'
olive_oil_df['Olives_Produced_kg_per_hectare_quantile_bins'] = pd.qcut(olive_oil_df['Olives_Produced_kg_per_hectare'], 10)

# Group by the new quantile bins and count the occurrences of each Soil_Type
# Then unstack to prepare for plotting and create a bar plot
soil_type_counts = olive_oil_df.groupby('Olives_Produced_kg_per_hectare_quantile_bins')['Soil_Type'].value_counts().unstack(fill_value=0)
soil_type_counts.plot.bar(figsize=(12, 6))
plt.title('Distribution of Soil_Type across Olives_Produced_kg_per_hectare Quantiles')
plt.xlabel('Olives_Produced_kg_per_hectare Quantile Bins')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
