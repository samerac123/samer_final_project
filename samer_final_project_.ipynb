### Correlation Heatmap

To understand the relationships between different variables, let's create a correlation heatmap. This will help us identify which features are strongly correlated with each other, especially with the target variables ('Oil_Yield_kg_per_hectare' and 'Olives_Produced_kg_per_hectare').

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Calculate the correlation matrix for numerical columns
correlation_matrix = olive_oil_df.select_dtypes(include=['number']).corr()

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# Draw the heatmap with seaborn
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)

plt.title('Correlation Matrix of Olive Oil Dataset Features', fontsize=16)
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
